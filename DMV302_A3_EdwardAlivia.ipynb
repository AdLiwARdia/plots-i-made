```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('/content/DMV302_Assessment_3_Dataset2.xlsx')

# labelled the columns to read better
dates = data['Date']
actual_avg_temp = data['Avg temp ']
actual_min_temp = data['Min temp']
actual_max_temp = data['Max temp']
avg_min_temp_since_1901 = data['avg temp since 1901']
avg_max_temp_since_1901 = data['Max temp since 1901']
lowest_temp_since_1901 = data['lowest temp since 1901']
highest_temp_since_1901 = data['Highest temp since 1901']
actual_rain = data['Rain amount']
avg_rain_since_1901 = data['AVG rain amount 1901']
highest_rain_since_1901 = data['highest rain 1901']

fig, weathertable = plt.subplots(figsize=(10, 6))

# placed labels to read better
weathertable.plot(dates, actual_avg_temp, label='Actual Average Temperature')
weathertable.plot(dates, actual_min_temp, label='Actual Minimum Temperature')
weathertable.plot(dates, actual_max_temp, label='Actual Maximum Temperature')
weathertable.plot(dates, avg_min_temp_since_1901, label='Average Min Temp Since 1901')
weathertable.plot(dates, avg_max_temp_since_1901, label='Average Max Temp Since 1901')
weathertable.plot(dates, lowest_temp_since_1901, label='Lowest Temp Since 1901')
weathertable.plot(dates, highest_temp_since_1901, label='Highest Temp Since 1901')
weathertable.plot(dates, actual_rain, label='Actual Amount of Rain')
weathertable.plot(dates, avg_rain_since_1901, label='Average Rain Since 1901')
weathertable.plot(dates, highest_rain_since_1901, label='Highest Rain Since 1901')

weathertable.set_xlabel('Date')
weathertable.set_ylabel('Value')

# added a title for plot
plt.title('Weather Data Over Time')

#added a legend for color coding
plt.legend()

plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# used xlsx as i am more comfortable with it
data = pd.read_excel('/content/DMV302_Assessment_3_Dataset1.xlsx')

# Extract the features for clustering (assuming X, Y, Z columns)
# this part took me a bit of time as i didnt notice that it was case sensitive and the labels in my excel werent upper case
X = data[['X', 'Y', 'Z']]

num_clusters = 3

# I am using k-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Get the cluster labels and cluster centers
labelplacer = kmeans.labels_
centers = kmeans.cluster_centers_

# puts shapes and color on clusters
markershapes = ['o', 's', '^']
coloring = ['red', 'green', 'blue']

fig = plt.figure(figsize=(8, 6))
bigcluster = fig.add_subplot(111, projection='3d')

# Plot the data points with color-coded clusters and different marker shapes
for i in range(num_clusters):
    cluster_points = X[labelplacer == i]
    bigcluster.scatter(cluster_points['X'], cluster_points['Y'], cluster_points['Z'],
               c=np.array([coloring[i]] * len(cluster_points)), marker=markershapes[i])


bigcluster.scatter(centers[:, 0], centers[:, 1], centers[:, 2], c='black', marker='x', s=100)

# place labels
bigcluster.set_xlabel('X')
bigcluster.set_ylabel('Y')
bigcluster.set_zlabel('Z')

plt.show()
```
